```python
import random

print('===== ROLADOR DE DADOS =====')

nome = input(f'Digite seu nome: ')
lados = int(input(f'\nQual o dado? D'))
quantidade = int(input(f'Quantos dados que rolar? '))

resultados = []

for i in range(quantidade):
  resultado = random.randint(1, lados)
  resultados.append(resultado)

resultados_texto = ', '.join(map(str, resultados))

print(f'\n{nome} rolou {quantidade}d{lados} e tirou:\n{resultados_texto}')




```
