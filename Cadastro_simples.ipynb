```python

print("=== Sistema de Cadastro ===")

cadastro = []

while True:

    print("\n1 - Adicionar nome")
    print("2 - Listar nomes")
    print("3 - Remover nome")
    print("4 - Sair")

    opcao = input("Escolha uma opção: ")

    if opcao == "1":
        nome = input("Digite o nome para cadastrar: ")
        cadastro.append(nome)
        print("Nome adicionado com sucesso!")

    elif opcao == "2":
        if len(cadastro) == 0:
            print("Nenhum nome cadastrado.")
        else:
            print("Nomes cadastrados:")
            for nome in cadastro:
                print("-", nome)

    elif opcao == "3":
        nome_remover = input("Digite o nome que deseja remover: ")
        if nome_remover in cadastro:
            cadastro.remove(nome_remover)
            print("Nome removido com sucesso!")
        else:
            print("Nome não encontrado.")

    elif opcao == "4":
        print("Encerrando o sistema...")
        break

    else:
        print("Opção inválida.")
```
